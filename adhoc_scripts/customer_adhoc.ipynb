```python
%sql
create or replace temp view customer_consolidated as 
select * from json.`/Volumes/gizmobox/landing/operational_data/customers/`
```

```python
%sql
select * from customer_consolidated where customer_name = 'John Smith' or customer_name = 'Andrew Taylor'
```

```python
%sql

select * from customer_consolidated 
where customer_id in (
    select customer_id
    from customer_consolidated
    group by customer_id
    having count(*)>1
)
order by customer_id
```

```python
%sql
select mst.* from customer_consolidated as mst
inner join (
    select customer_id, created_timestamp
    from customer_consolidated
    group by customer_id,created_timestamp
    having count(*)>1
) as chd
on mst.customer_id = chd.customer_id
order by mst.customer_id

```
