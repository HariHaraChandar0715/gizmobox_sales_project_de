```python
%sql
with cte_customers_data as (
  select * 
  from json.`/Volumes/gizmobox/landing/operational_data/customers/`
),
cte_duplicate_customers as (
  select
    customer_id,
    count(*) as cst_counts
  from cte_customers_data
  group by customer_id
),
cte_customer_duplicated_records as (
  select
    cst_src.customer_id,
    cst_src.created_timestamp as created_at,
    cst_src.email,
    cst_src.telephone,
    tgt_dup.cst_counts
  from cte_customers_data as cst_src
  left join cte_duplicate_customers tgt_dup
  on cst_src.customer_id = tgt_dup.customer_id
  where tgt_dup.cst_counts > 1
  order by cst_src.customer_id desc

)
select * from cte_customers_data;
```
